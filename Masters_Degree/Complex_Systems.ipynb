```python
# Programa que utiliza el método de Euler para
# graficar las ecuaciones de Lorenz

#Definir parámetros de EDO
delta_t=0.01 #Unidades de tiempo arbitrarias
t_final=100 #Unidades de tiempo arbritarias
#Calcular número de pasos
n_steps=int(t_final/delta_t) #int() para que sea entero
#Definir arreglo con tiempo en que se evalua la ecuación
t_array=np.linspace(0,t_final,n_steps)
#Definir la condición inicial
X_0=[10,10,10]

#Inicializar arreglo para almacenar las variables
X_array=np.zeros([n_steps,3])
#Intercambiar condición inicial en el arreglo
X_array[0,:]=X_0
#Definir parámetros de las ecuaciones
sigmma=1
r=80
b=1
#Integrar ecuación diferencial
for t in np.arange(n_steps-1):
  #Actualizar valor de Represores
  #Variable x
  X_array[t+1,0]=X_array[t,0]+delta_t*(sigmma*(X_array[t,1]-X_array[t,0]))
  #Variable y
  X_array[t+1,1]=X_array[t,1]+delta_t*(r*X_array[t,0]-X_array[t,1]-X_array[t,0]*X_array[t,2])
  #Variable z
  X_array[t+1,2]=X_array[t,2]+delta_t*(X_array[t,0]*X_array[t,1]-b*X_array[t,2])
```

Graficación

```python
plt.plot(t_array,X_array)
plt.legend(['x','y','z'])
plt.xlabel("tiempo")
plt.ylabel("Variables")
```

```python
#Importar librerias
import matplotlib.pyplot as plt
import numpy as np

# Inicializar las variables
X = []
Y = []

# Variación del parámetro de bifurcación
for r in np.linspace(0.7,4,100000):
    # Guarda el parámetro de bifurcación en la coordenada X
    X.append(r)
    # Genera un valor aleatorio para x, es la condición inicial
    x = np.random.random()
    # Llega al estado estacionario mediante iteraciones consecutivas
    for n in range(2000):
      x=(r*x)*(1-x)
    # Guarda el valor de x para cada valor del parametro
    Y.append(x)
# Grafica
plt.plot(X, Y, ls='', marker=',')
plt.xlabel('Parámetro de bifurcación')
plt.ylabel('Puntos de equlibrio')
plt.show()
```
