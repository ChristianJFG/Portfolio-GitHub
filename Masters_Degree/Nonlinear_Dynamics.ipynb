```python
# Programa que calcula el plano fase de un sistema de Edos
# 1D arrays
u=np.arange(-10,10,0.1)
v=np.arange(-10,10,0.1)

#Meshgrid
x,y=np.meshgrid(u,v)

#Assing vector directions
dx=-np.sin(x)**2 *np.sin(y)
dy=-2*np.sin(x)*np.cos(x)*np.cos(y)

#Depict illustration

plt.streamplot(x,y,dx,dy,density=3,linewidth=None,color='k')
plt.title('Plano fase')
plt.xlabel('$x$')
plt.ylabel('$y$')

#Show plot with grid
plt.grid()
plt.show()
```

```python
# Mapeo de henon
import numpy as np
import matplotlib.pyplot as plt

# Crear arrays
x0,y0 = [],[]
x1,y1 = [],[]
x2,y2 = [],[]
x3,y3 = [],[]
x4,y4 = [],[]

#Definir constantes
r = 0.3
a = 0
b = 0.4
x_equilibrio = -0.6
y_equilibrio = -0.6
n=8000 # Número de iteraciones

for i in np.arange(n):
  x0.append(x_equilibrio - r * np.cos(i))
  y0.append(y_equilibrio - r * np.sin(i))

for i in np.arange(n):
  x1.append(a - x0[i] ** 2 + b * y0[i])
  y1.append(x0[i])

for i in np.arange(n):
  x2.append(a - x1[i] ** 2 + b * y1[i])
  y2.append(x1[i])

for i in np.arange(n):
  x3.append(a - x2[i] ** 2 + b * y2[i])
  y3.append(x2[i])


for i in np.arange(n):
  x4.append(a - x3[i] ** 2 + b * y3[i])
  y4.append(x3[i])

# Graficar

plt.plot(x0,y0,'o',markersize = 0.09)
plt.plot(x1,y1,'o',markersize = 0.09)
plt.plot(x2,y2,'o',markersize = 0.09)
plt.plot(x3,y3,'o',markersize = 0.09)
plt.plot(x4,y4,'o',markersize = 0.09)
plt.xlabel('x')
plt.ylabel('y')

plt.xlim([-1.5,0])
plt.ylim([-1.5,0])

plt.show()
```
