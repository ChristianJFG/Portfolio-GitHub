```python
# Probabilidad Frecuentista
#Por Christian Jess Flores Gmez para Sistemas Complejos I
import random #libreria que genera nmeros aleatorios
import matplotlib.pyplot as plt #libreria para graficar
import numpy as np #libreria para calculo numrico
k=0
l=0 #Variables que se acumulan
num=500 # Nmero de tiros
y_1=[] #Vector de caras
y_2=[] #Vector de soles
for n in range(1,num+1): #primer ciclo for para los n lanzamientos
  for i in range(n-1): #segundo ciclo for
    volado=random.choice([0, 1]) #generacin de un nmero aleatorio entre 0 y 1
    if volado == 0: #Si sale cara...
      num_caras=k+1
      k=num_caras #Acumulacin de caras
    else:
      num_soles=l+1 #Si sale sol...
      l=num_soles #Acumulacin de soles
  y_1.append(k/n)
  y_2.append(l/n) #llenado del vector que almacena los datos de cada experimento
  k=0
  l=0 #Reiniciar las variables

tiros=np.linspace(1,n,n) #creacin de un vector del 1 a n
plt.plot(tiros,y_1,linewidth=0.3,label="Caras")
plt.plot(tiros,y_2,linewidth=0.3,label="Soles") #Graficacin
plt.xlabel("Tiradas")
plt.ylabel("Frecuencias relativas")
plt.legend()
```
